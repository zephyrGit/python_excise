```python
import pandas as pd
```

```python
f = open('一出好戏.txt',encoding='utf-8',errors='ignore')
```

```python
data = pd.read_csv(f,sep=',',header=None,encoding='utf-8',names=['date','nickName','city','rate','comment'])
```

```python
data
```

```python
city = data.groupby(['city'])
```

```python
city
```

```python
# 聚合函数
rate_group = city['rate']
```

```python
rate_group.mean()
```

```python
city.mean()
```

```python
# 平均分 数量
city_com = city['rate'].agg(['mean','count'])
```

```python
city_com
```

```python
# inplace 默认为False 
city_com.reset_index(inplace=True)
```

```python
city_com
```

```python
city_com['mean']=round(city_com['mean'],2)
```

```python
city_com['mean']
```

```python
city_com.shape
```

```python
data_map = [(city_com['city'][i], city_com['count'][i]) for i in range(0, city_com.shape[0])]
```

```python
data_map
```

```python
from pyecharts import Geo
```

```python
geo = Geo(
    "一出好戏城市评分", 
    "数据来源：猫眼",
    title_color="#fff",
    title_pos="center",
    width=1200,
    height=600,
    background_color="#404a59",
)
```

```python
geo
```

```python
attr, value = geo.cast(data_map)
geo.add(
    "",
    attr,
    value,
    visual_range=[0, 200],
    visual_text_color="#fff",
    symbol_size=15,
    is_visualmap=True,
)
geo.rander()
```
